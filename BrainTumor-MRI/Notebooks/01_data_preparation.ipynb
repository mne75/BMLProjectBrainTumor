```python
# Imports
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Paths to data folders
train_dir = "../data/Training"
test_dir = "../data/Testing"

# Image generators with rescaling
train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

# Load training and testing data
train_data = train_datagen.flow_from_directory(
    train_dir, 
    target_size=(150, 150),
    batch_size=32, 
    class_mode='binary',
    shuffle=True
    )

test_data = test_datagen.flow_from_directory(
    test_dir,
    target_size=(150, 150),
    batch_size=32, 
    class_mode='binary',
    shuffle=False
    )

# Print class labels and sample counts
print(" Class indices:", train_data.class_indices)
print(f"Total training samples: {train_data.samples}")
print(f"Total testing samples: {test_data.samples}")

# Visualise sample images
images, labels = next(train_data)
plt.figure(figsize=(12, 6))
for i in range(5):
    plt.subplot(1, 5, i+1)
    plt.imshow(images[i])
    plt.title("Tumor" if labels[i] == 1 else "No Tumor")
    plt.axis('off')
plt.suptitle("Sample Training Images")
plt.tight_layout()
plt.show()
```
